```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv("C:/Users/sai/Desktop/DSBDA/Assi3/Iris.csv")
df
```

```python
mean_sepal_length = df.loc[:, "SepalLengthCm"].mean()
print("Mean Sepal Length:", mean_sepal_length)

median_sepal_length = df.loc[:, "SepalLengthCm"].median()
print("Median Sepal Length:", median_sepal_length)

mode_sepal_length = df.loc[:, "SepalLengthCm"].mode()
print("Mode Sepal Length:", mode_sepal_length)

std_sepal_length = df.loc[:, "SepalLengthCm"].std()
print("Standard Deviation of Sepal Length:", std_sepal_length)

var_sepal_length = df.loc[:, "SepalLengthCm"].var()
print("Variance of Sepal Length:", var_sepal_length)

min_sepal_length = df.loc[:, "SepalLengthCm"].min()
print("Minimum Sepal Length:", min_sepal_length)

max_sepal_length = df.loc[:, "SepalLengthCm"].max()
print("Maximum Sepal Length:", max_sepal_length)

count_sepal_length = df.loc[:, "SepalLengthCm"].count()
print("Count of Sepal Length:", count_sepal_length)
```
