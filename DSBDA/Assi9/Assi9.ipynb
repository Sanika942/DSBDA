```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_data = sns.load_dataset("titanic")
```

```python
titanic_data.head(5)
```

```python
titanic_data.tail()
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data['age'].fillna(titanic_data['age'].mean(), inplace=True)
titanic_data['age'] = titanic_data['age'].astype(int)
```

```python
titanic_data['deck'].fillna('Unknown', inplace=True)
titanic_data['embarked'].fillna('Unknown', inplace=True)
titanic_data['embark_town'].fillna('Unknown', inplace=True)
```

```python
titanic_data.isnull().sum()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic_data, x='sex', y='age', hue='survived')
plt.title('Distribution of Age by Gender and Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')
# plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```
